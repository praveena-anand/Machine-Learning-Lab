**K MEANS CLUSTERING:**

**Image Compression:**

Import Required Libraries:

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
from PIL import Image
```

Load the Image:

```python
img = Image.open("/content/images.jpeg")
img = img.resize((396, 396))
img_np = np.array(img)
```

Reshape Image into 2D Array (pixels Ã— 3):

```python
pixels = img_np.reshape(-1, 3)
```

Apply K-Means Clustering (Choose K):

```python
K = 16
kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)
kmeans.fit(pixels)
```

Replace Each Pixel With its Cluster Center:

```python
compressed_pixels = kmeans.cluster_centers_[kmeans.labels_]
compressed_pixels = np.clip(compressed_pixels.astype('uint8'), 0, 255)
```

Reshape Back to Image Shape:

```python
compressed_img = compressed_pixels.reshape(img_np.shape)
```

Display Original and Compressed Images:

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img_np)
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(compressed_img)
plt.title(f"Compressed Image (K={K})")

plt.show()
```
