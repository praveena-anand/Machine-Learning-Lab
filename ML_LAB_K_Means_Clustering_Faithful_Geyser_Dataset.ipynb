**K MEANS CLUSTERING:**

**Faithful Geyser Dataset:**

Import Required Libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

Load the Old Faithful Geyser Dataset:

```python
df = sns.load_dataset("geyser")
```

Display First Few Rows:

```python
print(df.head())
```

Select Relevant Features:

```python
X = df[['duration', 'waiting']]
```

Standardize the Data:

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Apply K-Means Clustering with 2 Clusters:

```python
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

Display Cluster Centers (In Original Scale):

```python
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
print("\nCluster Centers (Duration, Waiting Time):")
print(centroids)
```

Plot the Clusters:

```python
plt.figure(figsize=(8,6))
plt.scatter(df['duration'], df['waiting'], c=df['Cluster'], cmap='rainbow', s=50)
plt.scatter(centroids[:,0], centroids[:,1], c='black', s=200, alpha=0.6, marker='X')
plt.title('K-Means Clustering of Old Faithful Geyser Eruptions')
plt.xlabel('Eruption Duration (minutes)')
plt.ylabel('Waiting Time (minutes)')
plt.show()
```
