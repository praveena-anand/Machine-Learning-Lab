**Random Dataset**

Importing Required Libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

Generate Random Dataset With Quadratic Decision Boundary:

```python
np.random.seed(42)
X = np.random.uniform(-3, 3, (200, 2))
y = (X[:, 1] > X[:, 0]**2).astype(int)
```

Train SVM Slassifier With RBF Kernel:

```python
clf = SVC(kernel='rbf', C=1, gamma=1)
clf.fit(X, y)
```

Plotting the Dataset and Decision Boundary:

```python
plt.figure(figsize=(8, 6))
```

Plot Data Points:

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
```

Plot Decision Boundary:

```python
xx, yy = np.meshgrid(np.linspace(-3, 3, 200), np.linspace(-3, 9, 200))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
plt.contourf(xx, yy, Z, alpha=0.2, cmap=plt.cm.coolwarm)
plt.contour(xx, yy, Z, colors='k', linewidths=0.5)

plt.title("SVM with Quadratic Boundary (RBF Kernel)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Train SVM Classifier With Linear Kernel:

```python
clf = SVC(kernel='linear', C=1)
clf.fit(X, y)
```

Plotting the Dataset and Decision Boundary:

```python
plt.figure(figsize=(8, 6))
```

Plot Data Points:

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
```

Plot Decision Boundary:

```python
xx, yy = np.meshgrid(np.linspace(-3, 3, 200), np.linspace(-3, 9, 200))
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

Draw Decision Boundary and Margins:

```python
plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='k')
plt.contour(xx, yy, Z, levels=[-1, 1], linewidths=1, linestyles='--', colors='k')

plt.title("SVM with Linear Kernel on Quadratic Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
